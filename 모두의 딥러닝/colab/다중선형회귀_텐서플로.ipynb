# 텐서플로에서 실행하는 선형 회귀 모델

## 1. 환경 준비

```python
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로의 케라스 API에서 필요한 함수 불러오기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## 2. 데이터 준비

```python
x = np.array([[2, 0], [4, 4], [6, 2], [8, 3]])
y = np.array([81, 93, 91, 97]) 
```

## 3. 모델 실행

```python
model = Sequential()

# 출력 값, 입력 변수, 분석 방법에 맞게끔 모델을 설정
model.add(Dense(1, input_dim=2, activation='linear'))
```

출력되는 값(=성적)이 하나씩이므로 Dense() 함수의 첫 번째 인자에 1이라고 설정한다.  
입력될 변수(=학습 시간, 과외 시간)가 두개이므로 input_dim은 2라고 설정한다.  
activation(활성화 함수)는 linear(선형 회귀)로 설정한다.

```python
# 오차 수정을 위해 경사 하강법(sgd)을, 오차의 정도를 판단하기 위해 평균 제곱 오차(mse)를 사용
model.compile(optimizer='sgd', loss='mse')
```

경사 하강법을 사용하기 위해 옵티마이저에 sgd라고 설정한다.  
손실 함수는 평균 제곱 오차를 사용할 것이므로 mse라고 설정한다.

```python
# 오차를 최소화하는 과정을 2000번 반복
model.fit(x, y, epochs=2000)
```

```python
import tensorflow as tf

# 임의의 학습 시간과 과외 시간을 집어넣어 점수를 예측하는 모델을 테스트

hour = 7
private_class = 4

input_data = tf.constant([[hour, private_class]])
prediction = model.predict(input_data)[0][0]

print("%.f시간을 공부하고 %.f시간의 과외를 받을 경우, 예상 점수는 %.02f점입니다" % (hour, private_class, prediction))
```
